# Environment settings

```python
import pandas as pd
import re
import numpy as np
import datetime
```

# Establish a working data frame

## Import the tabula CSV files

```python
# Import the raw data
df_pages1 = pd.read_csv('output_data/tabula_output/tabula-reports_first_pages_cropped.csv', header=None)
df_pages2 = pd.read_csv('output_data/tabula_output/tabula-reports_non_first_pages_cropped.csv', header=None)
```

```python
print(df_pages1.shape)
print(df_pages2.shape)
```

## Merge the different dataframes into one

```python
df = pd.concat([df_pages1, df_pages2]).reset_index(drop=True)

df.columns = ['original_date', 'update_date', 'county_city', 'originating_agency', 'jurisdictiont_tracking_num',
              'state_tracking_num', 'emac_federal_ma_num', 'subject', 'priority', 'worked_by', 'overall_status', 
              'assignment_status', 'assigned_to']
print(df.shape)
```

```python
df.head(10)
```

## Cleaning

Eliminate records that contain nothing but null values. (Again, we are keeping the oritinal index.)

```python
df = df.dropna(axis=0, how='all')

# Remember the number of rows of df at this point. We will need it to run consistency tests later.
original_lenght = len(df)

print(df.shape)
```

```python
# Break up df into two dataframes:
df_normal = df[df['original_date'].str.contains('^\d{2}\/\d{2}\/\d{4}.*', na=False)]
df_noise_values = df[~df['original_date'].str.contains('^\d{2}\/\d{2}\/\d{4}.*', na=False)]

# Consistency test
assert len(df_normal) + len(df_noise_values) == original_lenght
```

```python
df_noise_values
```

It seems clear that for a lot of the records in 'df_noise_values' where the first column contains noise, the values are shifted one column to ther right. Let's correct that.

```python
df_noise_values = df_noise_values.shift(periods=-1, axis='columns')
df_noise_values
```

Let's reunite the splitted dataframes into one.

```python
del(df)
df = pd.concat([df_normal, df_noise_values])
print(df.shape)

# Consistency test
assert len(df) == original_lenght
```

Have a look at the records in the restored dataframe that show null values in 'original_date' and see if we can safely eliminate them.

```python
df[df['original_date'].isna()]
```

They all look like noise. Away with them.

```python
df = df[~df['original_date'].isna()]
print(df.shape)

# Consistency test: Confirm that all the remainign 'orignal_date' values are proper dates.
assert df['original_date'].str.match('^\d{2}\/\d{2}\/\d{4}.*', na=False).all()
```

When we split the **df** into **df_normal** and **df_noise_values**, and then merged the last two back together, the index order of the new **df** was scrambled. Let's return things to place. (Remember that the index values represent individual orders.)

```python
df = df.sort_index(axis=0, inplace=False)
```

```python
df.head()
```

```python
df.tail()
```

## Dates

```python
df['original_dt'] = df['original_date'].str.extract('^(\d{2}\/\d{2}\/\d{4}).*$')
df['original_dt'] = pd.to_datetime(df['original_dt'])

df['update_dt'] = df['update_date'].str.extract('^(\d{2}\/\d{2}\/\d{4}).*$')
df['update_dt'] = pd.to_datetime(df['update_dt'])
```

# Item analysis

## Explosion by items

First, for each order (row), we create a list of the items requested in that order.

```python
df['subject_split'] = df['subject'].str.split('/') #('\r|,|/')
print(df.shape)
```

```python
df.head(10)
```

Right now we have one row per order. Expressed differently, each row represents one and only one order. 

Now, we will 'explode' each row according to the number of items in each order. So if order X requested Y different types of items, there will now be Y rows for that order, instead of only one.


```python
temp = df[~df['subject_split'].isna()]
temp_lenght = len(df[df['subject_split'].isna()])

num_items = 0
for subject in temp['subject_split']:
    num_items += len(subject)
    

print('The exploded dataframe should have exactly', num_items + temp_lenght, 'rows.')
```

```python
df = df.explode('subject_split')
df['subject_split'] = df['subject_split'].str.lower().str.strip()

print(df.shape)

# Consistency test
assert len(df) == num_items + temp_lenght
```

```python
df.head(15)
```

Eliminate rows where 'subject_split' resulted in an empty cell. First thing to know: how many records have an empty 'subject_split' field?

```python
df[df['subject_split'].str.match('^$', na=False)]['subject_split'].value_counts()
```

In the next line, we get rid of the records were the 'subject_split' fields shows up empty.

Notice we are not resetting the index. This is because each index of the df represents a unique order. So when we explode the df according to each item from each order, we want those expanded records keep that original index.mym

```python
df = df[~df['subject_split'].str.match('^$', na=False)]
print(df.shape)
```

Now that we have separated the order items into different rows, we can finanlly, eliminate all the carriage return characters (\r) embedded in practially all values in the df. (We could not do this before because they were used as a criteria to separate the items.)

```python
df = df.replace('\r', '', regex=True)
print(df.shape)
```

## Standardizing items

```python
df['subject_std'] = np.nan

df.loc[df['subject_split'].str.lower().str.match('^.*thermometer.*$', na=False), 'subject_std'] = 'Thermometers'
df.loc[df['subject_split'].str.lower().str.match('medical protective\requipment', na=False), 'subject_std'] = 'Protective Equipment'
df.loc[df['subject_split'].str.lower().str.match('medication', na=False), 'subject_std'] = 'Medication'
df.loc[df['subject_split'].str.lower().str.match('^.*swab.*$', na=False), 'subject_std'] = 'SWABS'
df.loc[df['subject_split'].str.lower().str.match('^.*soap.*$', na=False), 'subject_std'] = 'Soap'
df.loc[df['subject_split'].str.lower().str.match('^.*(face\sshield|shields?,\sface|shield,\rface|face\rshields).*$', na=False), 'subject_std'] = 'Face Shield'
df.loc[df['subject_split'].str.lower().str.match('^.*(gogg|glasses|eye\r?\s?protection).*$', na=False), 'subject_std'] = 'Goggles/Glasses'
df.loc[df['subject_split'].str.lower().str.match('ppe', na=False), 'subject_std'] = 'PPE'
df.loc[df['subject_split'].str.lower().str.match('^.*(\r?disinfect|\r?sanit|sanatizer|bleach|wipe).*$', na=False), 'subject_std'] = 'Disinfectant/Sanitizer'
df.loc[df['subject_split'].str.lower().str.match('^.*staffing.*$', na=False), 'subject_std'] = 'Staffing'
df.loc[df['subject_split'].str.lower().str.match('^.*glove.*$', na=False), 'subject_std'] = 'Gloves'
df.loc[df['subject_split'].str.lower().str.match('^.*gown.*$', na=False), 'subject_std'] = 'Gowns'
df.loc[df['subject_split'].str.lower().str.match('^.*mask.*$', na=False), 'subject_std'] = 'Masks'
```

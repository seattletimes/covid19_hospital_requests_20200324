## Comparing methods to calculate felony approval rates

#### Method 1: Average monthly approval rates by offense type

(**Manuel**) To calculate each DA's approval rate, we will:
1. Calculate the proportion of cases they approved each month (monthly approval rates,) broken down by offense
1. Average those monthly approval rates

```python
#Percentage of monthly cases appproved
dfp_app_alvarez = (dfp_intake_nonnull_app/dfp_intake_nonnull).loc[alv_start:alv_end]
dfp_app_foxx = (dfp_intake_nonnull_app/dfp_intake_nonnull).loc[foxx_start:foxx_end]
```

#### Method 2: Aggregate approval rates by offense type

(**Manuel**) So far we computed approval rates for each DA  by using the **average monthly approval rate** method: First, calculating their monthly approval rates by offense type and then averaging them.

Here, we use the **aggregate approval rate**. That is, we sum the total number of all the people whose cases were approved for prosecution throughout each DA's term (broken down by offense type) and then divide them by the sum of all people who were brought in for felony review throughout each DA's term
—to decide if their cases would be prosecuted or not—also broken down by offense type.

We are doing this to confirm the results don't materially change depending on the method we use to assess the proportion of **people** that ended up being sent  to prosecution by each DA.

```python
dfp_alv_all = dfp_intake_nonnull.loc[alv_start:alv_end]
dfp_alv_app = dfp_intake_nonnull_app.loc[alv_start:alv_end]

alv_total_approved = dfp_alv_app.sum()/ dfp_alv_all.sum()

delta = alv_total_approved.rename('alv_app_rate_monthly').to_frame().join(alv_mean_app)
delta['rate_diff'] = (delta['alv_app_rate_monthly'] - delta['Alvarez']).abs()

delta = delta.join(dfp_intake_nonnull.loc[alv_start:foxx_end].sum().rename('total_cases_alv').to_frame())
delta['people_diff'] = delta['rate_diff'] * delta['total_cases_alv']
```

```python
delta.sort_values(by = 'rate_diff', ascending=False).head(20)
```

```python
delta['people_diff'].sum()
```

The rate that shows the greatest difference when calculated with both methods is *Posession of Contraband in Penal Institution*. The difference is about nine percentage points.That is no small amount, of course, but on the other side, the number of people associated with those cases for both D.A. is 117, which amounts to a difference of roughly 10 people, arguably not extremely high.

Overall, calculating the approval rates with one method or another leads to a total difference of 296 people, not a vast number if we consider there were 164K people involved in approved cases.

(**Manuel**): One point worth noticing is that during Alvarez's tenure , the number of people involved in cases approved for felony prosecution started declining, and that trend continued under Foxx. Expressed differently, the downward trend in the number of people prosecuted each month did not start with Foxx.


To guage the effect Kim Foxx's influence had on the number of people diverted from Court, there are three schools of thought:

**Case 1**: Keep Alvarez' law enforcement activity consistent through Foxx's term. That is to say, assume Kim Foxx's influence does not effect law enforcement, and the number of cases seen by FRU is the same as it was in the 23 months prior to her election. 

**Case 2**: Simulate Alvarez' approval rate in Foxx's term, as law enforcement acts on her influence and policy changes.

**Case 3**: Isolate either DAs in their time periods and compare raw numbers of people diverted per offense.

These three considerations provide a difference of over 950 people being diverted from their case being heard in court.

**In the following code, we assume case 2: simulate Alvarez's approval rate in Foxx's term.**


Let's make the chart for our article

```python
#sum of approved cases by month (orange line below)
approved_cases = dfp_intake_nonnull_app.sum(axis=1).rename('approved_cases').to_frame()

#approved_case_rate by month/offense
approved_case_rate = (dfp_intake_nonnull_app/dfp_intake_nonnull)


#average monthly approval rate by offense during alvarez's term
alv_approved_case_rate = approved_case_rate.loc[alv_start:alv_end].mean().rename('xrate').to_frame().reset_index().set_index(offense_category_variable)

alv_approved_case_rate = alv_approved_case_rate.fillna(1)

# # #join alv's avg monthly approval rate by offense to monthly caseload by offense during Foxx's term
foxx = dfp_intake_nonnull.loc[foxx_start:foxx_end]
joined = foxx.transpose().join(alv_approved_case_rate)

# #multiply Foxx's monthly caseload by the avg monthly approval rate and sum across offense types
alv_during_foxx_approved = joined.multiply(joined["xrate"], axis=0).drop('xrate', axis=1).transpose().sum(axis=1).rename('alv_during_foxx_approve').to_frame()

# #merge actual cases approved during foxx's term with modeled cases approved during foxx's term by alvarez
merged = alv_during_foxx_approved.join(approved_cases)

# #get actual cases approved by alvarez during her term
alv_approved = dfp_intake_nonnull_app[alv_start:alv_end].sum(axis=1).rename('approved_cases').to_frame()

# #add null values for her approvals during foxx's term
alv_approved["alv_during_foxx_approve"] = np.nan

print(merged.size)
print(alv_approved.size)

#combine the two tables: foxx actual/alv modeled, alv actual/null values for alv modeled
merged = merged.append(alv_approved)

print(merged.size)

#sort by date
merged = merged.sort_index()

#plot
merged.plot()
```

First, we need to map down the overly abundant offense categories.

#### Disposition mapping

As we saw above, there are 36 different types of charge disposition. Those are too many to handle. In the following steps, we reclassify them to only four types.

```python
disposition_mapping = {
   #Plea Of Guilty
   "Plea Of Guilty":1, 
   "Plea of Guilty - Lesser Included":1, 
   "Plea of Guilty - Amended Charge":1, 
   "Plea of Guilty But Mentally Ill":1, 
   #Verdict Guilty
   "Verdict Guilty":2, 
   "Verdict Guilty - Lesser Included":2, 
   "Verdict Guilty - Amended Charge":2, 
   "Verdict Guilty But Mentally Ill":2, 
   #Finding Guilty
   "Finding Guilty":3, 
   "Finding Guilty - Lesser Included":3, 
   "Finding Guilty - Amended Charge":3, 
   #Verdict-Not Guilty
   "Verdict-Not Guilty":4, 
   #Finding Not Not Guilty
   "FNG":5, 
   "Finding Not Not Guilty":6, 
   "Finding Guilty But Mentally Ill":6, 
   "FNG Reason Insanity":6, 
   #Finding no probable cause
   "FNPC":7,
   #Nolle Prosecution
   "Nolle Prosecution":8, 
   "Nolle On Remand":8, 
   "Nolle Pros - Aonic":8,
   #Case Dismissed                       
   "Case Dismissed":9, 
   #Charge Rejected
   "Charge Rejected":9, 
   "Charge Reversed":9, 
   "Charge Vacated":9, 
   "Withdrawn":9, 
   #Bond forfeiture warrant
   "BFW":10, 
   #Death suggested
   "Death Suggested-Cause Abated":11, 
   "SOL":12, 
   "SOLW":12, 
   "Superseded by Indictment":13, 
   "Mistrial Declared":13, 
   "Transferred - Misd Crt":13, 
   "WOWI":13, 
   "Sexually Dangerous Person":13, 
   "Hold Pending Interlocutory":13, 
   "TRAN-FAI":13}

disposition_mapping_top = {
   1:"Plea of guilty, to amended, to lesser included, or but mentally ill", 
   2:"Verdict guilty", 
   3:"Finding guilty", 
   4:"Verdict not guilty", 
   5:"Finding not guilty", 
   6:"FNG reason insanity, finding not not guilty", 
   7:"Finding no probable cause", 
   8:"Nolle Prosequi", 
   9:"Case dismissed", 
   10:"Bond forfeiture warrant", 
   11:"Death suggested", 
   12:"Stricken with leave to reinstate", 
   13:"Other"}

disposition_mapping_final = {
   # Finding or verdict of guilty
   "Verdict guilty":"Finding or verdict of guilty", 
   "Finding guilty":"Finding or verdict of guilty", 
   "Plea of guilty, to amended, to lesser included, or but mentally ill":"Finding or verdict of guilty", 
   # Finding or verdict of not guilty
   "Verdict not guilty":"Finding or verdict of not guilty", 
   "Finding not guilty":"Finding or verdict of not guilty", 
   "FNG reason insanity, finding not not guilty":"Finding or verdict of not guilty", 
   # Case dismissed or nolle pros
   "Finding no probable cause":"Case dismissed or nolle pros", 
   "Nolle Prosequi":"Case dismissed or nolle pros", 
   "Case dismissed":"Case dismissed or nolle pros", 
   "Stricken with leave to reinstate":"Case dismissed or nolle pros", 
   # Other
   "Bond forfeiture warrant":"Other", 
   "Death suggested":"Other", 
   "Other.":"Other"}

disposition_mapping_final_num = {
    "Finding or verdict of guilty":3, 
    "Finding or verdict of not guilty":2, 
    "Case dismissed or nolle pros":1, 
    "Other":0}

disposition_mapping_final_num_reverse = {
    3:"Finding or verdict of guilty", 
    2:"Finding or verdict of not guilty", 
    1:"Case dismissed or nolle pros", 
    0:"Other"}
```

We now create new columns to carry out the mapping.

```python
df_dispo["charge_disposition_agg"] = df_dispo['charge_disposition'].map(disposition_mapping)
df_dispo["charge_disposition_top"] = df_dispo['charge_disposition_agg'].map(disposition_mapping_top)
df_dispo["charge_disposition_final"] = df_dispo['charge_disposition_top'].map(disposition_mapping_final)
df_dispo["disposition_mapping_final_num"] = df_dispo['charge_disposition_final'].map(disposition_mapping_final_num)
```

```python
print('Ultimately, what this whole mapping is doing is reducing the', df_dispo['charge_disposition'].nunique(), 'classifications from charge_disposition to only', df_dispo['charge_disposition_final'].nunique(), '.')
print('\n')
print(df_dispo['charge_disposition'].value_counts(dropna=False))
print('\n')
print(df_dispo['charge_disposition_final'].value_counts(dropna=False))
print('\n')

#Consistency test
df_dispo['charge_disposition'].value_counts(dropna=False).sum() == df_dispo['charge_disposition_final'].value_counts(dropna=False).sum()
```

#### Delta

We will now calculate the hypotetical number of people who would have been prosecuted if Foxx had followed the average prosecution rates shown by Alvarez.

```python
#Total people involved in narcotics cases
alv_narc_totals = df_dispo_stemen_narcotics_deduped_pivot[alv_start:alv_end].sum().sum()
foxx_narc_totals = df_dispo_stemen_narcotics_deduped_pivot[foxx_start:foxx_end].sum().sum()

#Total people involved in narcotics cases that were dismissed
alv_narc_dismissed_totals = df_dispo_stemen_narcotics_deduped_pivot[alv_start:alv_end]['Case dismissed or nolle pros'].sum()
foxx_narc_dismissed_totals = df_dispo_stemen_narcotics_deduped_pivot[foxx_start:foxx_end]['Case dismissed or nolle pros'].sum()

#Dismiss rates
alv_narc_dismissed_rate = alv_narc_dismissed_totals / alv_narc_totals
foxx_narc_dismissed_rate = foxx_narc_dismissed_totals / foxx_narc_totals

print('alv_narc_totals =', alv_narc_totals)
print('alv_narc_dismissed_totals', alv_narc_dismissed_totals)
print('alv_narc_dismissed_rate', alv_narc_dismissed_rate)
print('\n')
print('foxx_narc_totals =', foxx_narc_totals)
print('foxx_narc_dismissed_totals', foxx_narc_dismissed_totals)
print('foxx_narc_dismissed_rate', foxx_narc_dismissed_rate)
```

```python
#multiply alvarez's dismissal rate by Foxx's total narcotics case load
alv_modeled = alv_narc_dismissed_rate * foxx_narc_totals
alv_modeled
```

### Calculating dismissal rates, accounting for felony class

From last section, we know that the dismissal rate has changed by Class severity under Foxx (especially more-severe cases).

Because of that, we want to account for cases that have mulitple felony classes. For example, how should we treat a Narcotics case with Felony Class 2 and Class 4 charges? 

What we'll do is look at the dismissal rate for every combination of felony charges and use that rate in our model.

```python
#remember the offense class mapping we set up earlier:
class_mapping
```

```python
#map felony classes
df_dispo_stemen_narcotics["class_mapped"] = df_dispo_stemen_narcotics['class'].map(class_mapping)
```

```python
print(df_dispo_stemen_narcotics.shape)
```

```python
#group cases by their max disposition date and de-duplicate multiple charges under one felony class
df_dispo_narcotics_grouped_by_charge_class = df_dispo_stemen_narcotics.groupby(['class_mapped','case_participant_id'], sort=False)['dispo_date'].max().to_frame().reset_index()

#sort by case and felony class
df_dispo_narcotics_grouped_by_charge_class = df_dispo_narcotics_grouped_by_charge_class.sort_values(['case_participant_id', 'class_mapped'], ascending=[False, False]).reset_index()
```

```python
print(df_dispo_narcotics_grouped_by_charge_class.shape)
df_dispo_narcotics_grouped_by_charge_class.head(10)
```

Concatanate the felony classes so we can then group them together to approximate the dismissal rate.

```python
# First, convert 'class_mapped' into a string.
df_dispo_narcotics_grouped_by_charge_class['class_mapped'] = df_dispo_narcotics_grouped_by_charge_class['class_mapped'].apply(str)
# Now concatenate them into a sincle cell for each participant/date
df_dispo_narcotics_grouped_by_charge_class = df_dispo_narcotics_grouped_by_charge_class.groupby(['dispo_date','case_participant_id']).agg({'class_mapped':'-'.join}).reset_index().set_index('case_participant_id')
```

```python
print(df_dispo_narcotics_grouped_by_charge_class.shape)
df_dispo_narcotics_grouped_by_charge_class.head(10)
#Note that this de-dupes for many chrages in one class (so we're not acocunting for charge counts on cases
```

We'll join with our previous calculation of the maximum case disposition for narcotics-only charges on a case.

```python
# Remembering the mapping:
disposition_mapping_final_num_reverse
```

```python
#remembering the DF that it will be joined with:
print(df_dispo_stemen_narcotics_deduped.shape)
df_dispo_stemen_narcotics_deduped.head(10)
```

```python
print(df_dispo_narcotics_grouped_by_charge_class_joined.shape)
df_dispo_narcotics_grouped_by_charge_class_joined.head(10)
```

Now we'll calculate dismissal rates for each charge permutation for both DAs.

```python
# Separate by DA
alv_charges = df_dispo_narcotics_grouped_by_charge_class_joined.set_index('dispo_date')[alv_start:alv_end]
foxx_charges = df_dispo_narcotics_grouped_by_charge_class_joined.set_index('dispo_date')[foxx_start:foxx_end]

# Mapping
alv_charges['disposition_mapping_final_num'] = alv_charges['disposition_mapping_final_num'].map(disposition_mapping_final_num_reverse)
foxx_charges['disposition_mapping_final_num'] = foxx_charges['disposition_mapping_final_num'].map(disposition_mapping_final_num_reverse)

# Count the number of of outcomes by each felony class string
alv_charges_pivot = alv_charges.pivot_table(columns='disposition_mapping_final_num', index='class_mapped', values='case_participant_id', aggfunc='count', fill_value = 0)
foxx_charges_pivot = foxx_charges.pivot_table(columns='disposition_mapping_final_num', index='class_mapped', values='case_participant_id', aggfunc='count', fill_value = 0)

# Compute rates
alv_charges_pivot = alv_charges_pivot.div(alv_charges_pivot.sum(axis=1), axis=0)
foxx_charges_pivot = foxx_charges_pivot.div(foxx_charges_pivot.sum(axis=1), axis=0)
```

```python
foxx_charges_pivot = foxx_charges.pivot_table(columns='disposition_mapping_final_num', index='class_mapped', values='case_participant_id', aggfunc='count', fill_value = 0).sum(axis=1).to_frame().rename(columns={0:'total_foxx'})
foxx_charges_pivot.head()
```

```python
# We'll join alvarez's average dismissal rate by felony class string with Foxx's total charge count by felony class string.
# Since there's not a perfect overlap of felony class strings, we'll give Alvarez the benefit of the doubt and assume those are dismissed -- therefore using fillna(1)
foxx_charges_pivot = foxx_charges_pivot.join(alv_charges_pivot).fillna(1)
```
